```python
from dateutil import parser
import pandas as pd

# Data viz packages
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# NLP
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('stopwords')
nltk.download('punkt')
from wordcloud import WordCloud
```

## Reading the video file as input

```python
video_df = pd.read_csv('video_details.csv')
```

```python
video_df.head()
```

## Best and Worst performing videos

Here, we want to know which 10 videos received the highest number of views. 

```python
sns.set_style("darkgrid") # Setting the style of the grid
```

```python
plt.figure(figsize=(8, 6))
ax = sns.barplot(x = 'title', y = 'viewCount', data = video_df.sort_values('viewCount', ascending=False)[0:9])
plot = ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos:'{:,.0f}'.format(x/1000) + 'K'))
plt.xlabel("Video")  # Label for the x-axis
plt.ylabel("View Count")   # Label for the y-axis
plt.title (" Top 10 high performing videos")

plt.show()
```

From this plot, we see that the video that received the highest views is the 'Mutton Kosha' video.

```python
plt.figure(figsize=(8, 6))
ax = sns.barplot(x = 'title', y = 'viewCount', data = video_df.sort_values('viewCount', ascending=True)[0:9])
plot = ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos:'{:,.0f}'.format(x/1000) + 'K'))
plt.xlabel("Video")  # Label for the x-axis
plt.ylabel("View Count")   # Label for the y-axis
plt.title (" Lowest performing videos")

plt.show()
```

## View distribution

```python
sns.histplot(data=video_df["viewCount"])
# Add labels and title
plt.xlabel(" View Counts")
plt.ylabel("Number of videos")

plt.title("View distribution")
# Show the plot
plt.show()
```

There are a few videos that have over 6 million views, for example, the 'Mutton Kosha' video above. Majority of their videos however are around 100,000 views. 

## Comments distribution

```python
sns.histplot(data=video_df["commentCount"])
# Add labels and title
plt.xlabel(" Number of Comments")
plt.ylabel("Number of videos")
plt.title("Comment distribution")
# Show the plot
plt.show()
```

## When are most videos published?

```python
## Published at
df = video_df.groupby(video_df["publishedDay"])['video_id'].count()
```

```python
 sns.barplot(data=df.sort_values())
# Add labels and title
plt.xlabel(" Day of Week ")
plt.ylabel("Count of Videos")
plt.title(" Count of videos published each day of week")
# Show the plot
plt.show()
```

Fridays seem to be the day when most videos are posted, 209 out of 292, and I think it makes sense considering people can try out new recipes over the weekend.

```python
video_df.loc[video_df['publishedDay'] == 'Wednesday']
```

Wednesday videos, the recent ones are shorts, except one full-length video, which was posted 3 years back. Since then , the posting strategy has evolved.

## Average Video Duration

```python
sns.histplot(data = video_df, x = 'durationSecs', bins=30)
# Add labels and title
plt.xlabel(" Video Duration (secs)")
plt.ylabel("Number of videos")
plt.title("Average duration of videos")
# Show the plot
plt.show()
```

There are few videos that are very long, but majority of the videos are around 10 mins long.

## Relation between view count and like count, comments, and duration

```python
#fig, ax = plt.subplots(1,2)
data = video_df[['viewCount','commentCount','likeCount','durationSecs']]
sns.pairplot(data)
```

Videos with higher view count clearly have higher likes and comments counts as well, and vice versa. I was also interested to learn whether longer videos garner lesser views or not. It is evident that that most popular videos, the ones with highest view or likes or comments were not necessarily the longest ones. In fact, for very long videos, the view count is on the lower side.

## Word Cloud of video titles

```python
stop_words = set(stopwords.words('english'))
video_df['title_no_stopwords'] = video_df['title'].apply(lambda x: [item for item in str(x).split() if item not in stop_words])

all_words = list([a for b in video_df['title_no_stopwords'].tolist() for a in b])
all_words_str = ' '.join(all_words) 

def plot_cloud(wordcloud):
    plt.figure(figsize=(30, 20))
    plt.imshow(wordcloud) 
    plt.axis("off");

wordcloud = WordCloud(width = 2000, height = 1000, random_state=1, background_color='black', 
                      colormap='viridis', collocations=False).generate(all_words_str)
plot_cloud(wordcloud)
```

Can there be anything more quintessentially Bengali and food than this? The word cloud shows everything that the channel is so loved for.
