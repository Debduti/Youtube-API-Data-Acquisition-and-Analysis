```python
def get_channel_stats(youtube,channel_ids): 

    all_items = []    
    request = youtube.channels().list(
        part="snippet,contentDetails,statistics",
        id=','.join(channel_ids)
        )
    response = request.execute()
       
    for item in response['items']:

        data = {'channelname':item['snippet']['title'],
               'views':item['statistics']['viewCount'],
               'subscribers':item['statistics']['subscriberCount'],
               'total_videos': item['statistics']['videoCount'],
               'playlist_id':item['contentDetails']['relatedPlaylists']['uploads']
                }
        
        all_items.append(data)

    return pd.DataFrame(all_items)       
    
    
    
```

```python
def get_video_ids(youtube,channel_ids):

    video_ids = []   
    next_page_token = ''
    while next_page_token is not None:
        request = youtube.playlistItems().list(
                part="snippet,contentDetails",
                playlistId="UUYLAf7WuBy6y64RewvwGjoQ",
                maxResults =50,
                pageToken = next_page_token
                )
        response = request.execute()    
    
        for item in response['items']:
                      
            video_ids.append(item['contentDetails']['videoId'])
        
        next_page_token = response.get('nextPageToken')
        
    
    return video_ids
    
```

```python
def get_video_details(youtube,video_ids):
    all_video_info =[]
    
    for i in range(0,len(video_ids),50): #6 iters
        request = youtube.videos().list(
            part="snippet,contentDetails,statistics",
            id=','.join(video_ids[i:i+50])
            )
        response = request.execute()        
        
        for video in response['items']: # 50 iters
        
            interesting_stats = {'snippet':['channelTitle','title','publishedAt','description'],
                                'contentDetails':['duration','definition','caption'],
                                'statistics':['viewCount','likeCount','favoriteCount','commentCount']
                                }
            video_info = {}
            video_info['video_id'] = video['id']
                  
            for k in interesting_stats.keys(): # 3 iters
                for v in interesting_stats[k]: # 4 iters
                    try:
                        video_info[v] =video[k][v]
                    except:
                        video_info[v] = None
        
            all_video_info.append(video_info)
               
    return pd.DataFrame(all_video_info)
    
    
    
```

```python
def get_all_comments(youtube,video_ids):
    all_comments =[]
    for video_id in video_ids:
        request = youtube.commentThreads().list(
                part="snippet,replies",
                videoId= video_id
              )
       
        try:    
            response = request.execute()
        except:
            response = ''     
       
        try:
            for item in response['items']:            
                comments_in_video_info = {'video_id':video_id,
                                      'comment':item['snippet']['topLevelComment']['snippet']['textOriginal']
                                      }
                all_comments.append(comments_in_video_info)
        except:
            pass
        
    return pd.DataFrame(all_comments)
            

```

```python
# We need to convert all Count columns to numeric

numeric_cols = ['viewCount', 'likeCount', 'favoriteCount', 'commentCount']
video_df[numeric_cols] = video_df[numeric_cols].apply(pd.to_numeric,axis =1)
```

```python
# Find out day of week from publishedAt

video_df['publishedAt'] = video_df['publishedAt'].apply(lambda x: parser.parse(x))
video_df['publishedDay'] = video_df['publishedAt'].apply(lambda x: x.strftime('%A'))

```

```python
video_df.head()
```

```python
# Calculate duration of videos in secs
import re

def iso_duration_to_seconds(duration_str):
    # Regular expression to extract hours, minutes, and seconds
    pattern = re.compile(r'P(?:T(?:(\d+)H)?(?:(\d+)M)?(?:(\d+)S)?)?')

    # Use regular expression to find matches in the duration string
    match = pattern.match(duration_str)

    if not match:
        raise ValueError("Invalid ISO 8361 duration format")

    # Extract hours, minutes, and seconds from the match
    hours = int(match.group(1)) if match.group(1) else 0
    minutes = int(match.group(2)) if match.group(2) else 0
    seconds = int(match.group(3)) if match.group(3) else 0

    # Calculate total seconds
    total_seconds = hours * 3600 + minutes * 60 + seconds

    return total_seconds
```

```python
video_df['durationSecs'] = video_df['duration'].apply(lambda x: iso_duration_to_seconds(x))
```

```python
#Clean all comments not in english

from langdetect import detect_langs


def is_english(comment):
    try:
        langs = detect_langs(comment)
        for lang in langs:
            if lang.lang == 'en':
                return 'True'
        return 'False'
    except:
        return 'False'



```

```python
# Apply the function to create a new column 'english_comments'
all_comments_df['english_comments'] = all_comments_df['comment'].apply(is_english)
```

```python
all_comments_df = all_comments_df[all_comments_df['english_comments'] == 'True'].reindex()
```
